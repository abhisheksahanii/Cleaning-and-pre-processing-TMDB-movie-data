I need a python file coded which preprocesses and cleans the TMDB movie data.
Following is required to be coded in python:
1. Take top 200 of rows in dataset with highest revenue value and put into dataframe top200
2. With the new dataframe top200, for each row that has multiple director names in the 'director' column, delete that row and add back a copy of that row but each of the new rows will contain one director each.  So if original 'director' column has 'Adam | Peter | Linda', there should be three copies of the original row, one row containing director = 'Adam', second copy of the row has 'Peter' as director and 'Linda' as director in the third copy of the row.
3. With the same dataframe top200, expand the same as #2 but for the column of genre, cast, and production studio

```python
import pandas as pd
```

```python
tmdb= pd.read_csv('tmdbmovies.csv')
```

```python
tmdb.head(5)
```

```python
top200 = tmdb.sort_values(by=['revenue'],ascending=False).head(200) #1. Take top 200 of rows in dataset with highest revenue value and put into dataframe top200
```

```python
top200 #printing top200
```

```python
top200['director']=top200.director.str.split("|") 
#2. With the new dataframe top200, for each row that has multiple director names in the 'director' column, delete that row and add back a copy of that row but each of the new rows will 
# contain one director each.  So if original 'director' column has 'Adam | Peter | Linda', there should be three copies of the original row, one row containing director = 'Adam', second copy of the row has 'Peter' as director and 'Linda' as director in the third copy of the row.
```

```python
top200_demo1=top200.explode('director')
```

```python
top200_demo1
```

```python
top200_demo1['cast']=top200_demo1.cast.str.split("|")
top200_demo2=top200_demo1.explode('cast')
```

```python
top200_demo2
```

#3.With the same dataframe top200, expand the same as #2 but for the column of genre, cast, and production studio

```python
top200_demo2['genres']=top200_demo2.cast.str.split("|") 
top200_demo3 = top200_demo2.explode('genres')
```

```python
top200_demo3
```

```python
top200_demo3['production_companies'] = top200_demo3.production_companies.str.split("|")
top200_demo4 = top200_demo3.explode('production_companies')
```

```python

top200_demo4.to_csv('tmdb_extended_Data.csv',index=False)
```
